```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.font_manager import FontProperties
```

```python
import matplotlib.font_manager

# Get a list of available font names
font_names = [f.name for f in matplotlib.font_manager.fontManager.ttflist]

# Print the list
print(font_names)

for font in font_names:
    if 'Arial' in font:
        print(font)
```

```python
# Create a new figure
fig = plt.figure(figsize=(15, 10))

# Create a GeoAxes in the tile's projection
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson())

# Make the map global rather than have it zoom in to the extents of any plotted data
ax.set_global()

# Add map features
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Define the locations
locations = [
    ("Taal Volcano", 14.0020, 120.9934),
    ("Ridgecrest Earthquake", 35.6225, -117.6709),
    ("San Quintin Glacier", -46.8625, -73.8491),
    ("Tien Shan Glacier", 42.0000, 80.0000)
]

addon_locations = [("Türkiye Earthquake", 37.166, 37.032)]
addon_locations2 = [("Mauna Loa Volcano", 19.479444, -155.602778)]

train_location = [("Alaska's Glaciers", 61.0, -147.5)]

# Set up Arial font
# font_prop = FontProperties(family='Arial Black')
font_prop = FontProperties(family='Arial Narrow',weight='bold')

# Plot each location
for name, lat, lon in locations:
    ax.plot(lon, lat, 'ro', markersize=10, transform=ccrs.Geodetic())
    # Adjust the longitude slightly to create a gap
    text_lon = lon + 20  # Adjust this value to increase/decrease the gap
    text_lat = lat + 5.8  # Adjust this value to increase/decrease the gap
    ax.text(text_lon, text_lat, name, fontsize=12, ha='right', va='center',
            transform=ccrs.Geodetic(), fontproperties=font_prop, bbox=dict(facecolor='white', edgecolor='none', alpha=0.6))
    
for name, lat, lon in addon_locations:
    ax.plot(lon, lat, 'ro', markersize=10, transform=ccrs.Geodetic())
    # Adjust the longitude slightly to create a gap
    text_lon = lon + 15  # Adjust this value to increase/decrease the gap
    text_lat = lat - 6  # Adjust this value to increase/decrease the gap
    ax.text(text_lon, text_lat, name, fontsize=12, ha='right', va='center',
            transform=ccrs.Geodetic(), fontproperties=font_prop, bbox=dict(facecolor='white', edgecolor='none', alpha=0.6))
    
for name, lat, lon in addon_locations2:
    ax.plot(lon, lat, 'ro', markersize=10, transform=ccrs.Geodetic())
    # Adjust the longitude slightly to create a gap
    text_lon = lon + 20  # Adjust this value to increase/decrease the gap
    text_lat = lat - 6.5  # Adjust this value to increase/decrease the gap
    ax.text(text_lon, text_lat, name, fontsize=12, ha='right', va='center',
            transform=ccrs.Geodetic(), fontproperties=font_prop, bbox=dict(facecolor='white', edgecolor='none', alpha=0.6))
    
for name, lat, lon in train_location:
    ax.plot(lon, lat, 'r*', markersize=15, transform=ccrs.Geodetic())
    # Adjust the longitude slightly to create a gap
    text_lon = lon + 40  # Adjust this value to increase/decrease the gap
    text_lat = lat - 6.4  # Adjust this value to increase/decrease the gap
    ax.text(text_lon, text_lat, name, fontsize=12, ha='right', va='center',
            transform=ccrs.Geodetic(), fontproperties=font_prop, bbox=dict(facecolor='white', edgecolor='none', alpha=0.6))

# # Add a title
# plt.title("Notable Locations Around the World", fontsize=16)

# Save the figure
plt.savefig('MKOT_places_around_the_woruld', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
